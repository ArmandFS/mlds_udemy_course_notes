```python
import numpy as np
import pandas as pd
```

```python
#get data and data references
hotels = pd.read_csv("hotel_booking_data.csv")
hotels
```

```python
#task 1: How many rows are there?
#with shape, can see how many items are remaining
#hotels.shape[1] gives number of columns
hotels.shape[0]
```

```python
#task 2: Is there any missing data? If so, which column has the most missing data?
#isnull to find empty data
#sum it to find missing data
hotels.isnull().sum()
```

```python
#Task 3: Drop the company column from the dataset.
hotels = hotels.drop('company',axis=1)
```

```python
#task 4: What are the top 5 most common country codes in the dataset.
#call values counts to sum values, and then slice to get top 5
hotels['country'].value_counts()[:5]
```

```python
#task 5: what is the name of the person who paid the highest ADR (average daily rate)? How much was their ADR?
#iloc 0 is to grab the first row
hotels.sort_values('adr', ascending=False)[['adr', 'name']].iloc[0]
```

```python
#Task 6: The adr is the average daily rate for a person's stay at the hotel. What is the mean adr across al the hotel stays in the dataset.
hotels['adr'].mean()
```

```python
#task 7. What is the average (mean) number of nights for a stay across the entire dataset?
#combine the 2 night columns
#you can just plus them
hotels['total_stay_nights'] =  hotels['stays_in_weekend_nights']+ hotels['stays_in_week_nights']
hotels['total_stay_nights'].mean()
```

```python
#task 8. What is the average total cost for stay in the dataset? Not average daily cost, but total stay cost. 
hotels['total_spend'] = hotels['adr'] * hotels['total_stay_nights'] 
hotels['total_spend'].mean()
```

```python
#task 9. What are the names and emails of people who made 5 Special Requests?
hotels[hotels['total_of_special_requests'] == 5][['name', 'email']]
```

```python
#task 10. What percentage of hotel stays were classified as "repeat guests"?
repeat_guest = sum(hotels['is_repeated_guest'] == 1) 
percent_repeat = 100 * (repeat_guest/len(hotels))
percent_repeat
```

```python
#Task 11. What are the top 5 most common last name in the dataset
#split the data in the names column
#get the last item in the list and then sort it
hotels['name'].apply(lambda name: name.split()[-1]).value_counts()[:5]
```

```python
#Task 12. What are the top 3 most common area code in the phone numbers?
hotels['phone-number'].apply(lambda num: num[:3]).value_counts()[:3]
```

```python
#Task 13. How many arrivals took place between the 1st and the 15th of the month?
hotels['arrival_date_day_of_month'].apply(lambda day: day in range(1,16)).sum()
```
