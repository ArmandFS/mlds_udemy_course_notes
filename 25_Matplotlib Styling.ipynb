### 25 - Legend Styling

```python
#import the matplotlib library and other ds libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#create data
a = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(a, a, label = 'A to A plot')
ax.plot(a,a**2, label = 'A vs A^2 plot')
ax.legend(loc=(1.1, 0.5))
```

### Styling, Colors, Markers etc.

```python
#create data for styling
x = np.linspace(0,11,10)
```

```python
fig = plt.figure(figsize=(12,8),dpi=80)
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, color='#3d90c4', label = 'X to X', lw = 3)
ax.plot(x, x*2, color='red', label = 'X times 2', lw = 4, ls="dashed")
ax.plot(x, x*3, color='#852fd6', label = 'X times 3', lw = 5)
ax.plot(x, x*4, color='#c73a9d', label = 'X times 4', lw = 6, ls="dashdot")
ax.plot(x, x*5, color='#38d66a', label = 'X times 5', lw = 7)
ax.plot(x, x*6, color='#dbd000', label = 'X times 6', lw = 8, ls="dotted")
ax.plot(x, x*7, color='#ff6803', label = 'X times 7', lw = 1, marker='o', markersize=10)
ax.plot(x, x*8, color='black', label = 'X times 8', lw = 1, marker='x', markersize=10)

ax.legend(loc="upper center", prop={'size': 12})
```

```python
# make a unique marker
fig = plt.figure(figsize=(12,8),dpi=70)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, color='purple', label = 'X to X', lw = 3, marker='s', ls='--', ms=10, markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
