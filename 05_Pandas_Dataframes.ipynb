#### Working with Columns and Data Retrieval

```python
#get column data
df1['total_bill']
```

```python
#list column names
df1[['total_bill', 'tip','size']]
```

```python
#calculations
df1['tip'] + df1['total_bill']
```

```python
100 * (df1['tip'] / df1['total_bill'])
```

```python
#create a new column
df1['tip_percentage'] = 100 * df1['tip'] / df1['total_bill']
df1.head()
```

```python
#create (overwrite) price per person column
df1['price_per_person'] = df1['total_bill'] / df1['size']
df1.head()
```

```python
#remove columns
df1.drop('tip_percentage', axis=1)
```
