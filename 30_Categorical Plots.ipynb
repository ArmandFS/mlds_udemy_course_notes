### 30 - Categorical Plots


```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Use same dataset
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
df['division'].value_counts()
```

```python
#create a countplot
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='division')
```

```python
#add a hue to the countplot
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='level of education', hue='division', palette='Set1')
```

```python
#create a barplot
#to visualize mean
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='level of education', y='salary', estimator= np.mean, errorbar='sd', hue='division')

#set the legend outside of the plot
plt.legend(bbox_to_anchor=(1.05, 1))
```
