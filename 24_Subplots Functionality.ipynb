### 24 - Subplots

```python
#import the matplotlib library and other ds libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#create data
a = np.linspace(0,10,11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
x
```

```python
y
```

```python
#new data
c = np.linspace(0,9,11)
c
```

```python
d = c ** 3
d
```

```python
e = np.arange(0,8)
f = 3 * e
```

```python
e
```

```python
f
```

```python
#create subplots with 1 row 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
#expand by 2 rows 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

```python
#2nd subplot
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(a,b)
axes[0][1].plot(c,d)
axes[1][0].plot(e,f)
axes[1][1].plot(x,y)
```

```python
#2nd subplot with titles

fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(a,b)
axes[0][0].set_title('Title 0 0')


axes[0][1].plot(c,d)
axes[0][1].set_title('Title 0 1')

axes[1][0].plot(e,f)
axes[1][0].set_title('Title 1 0')

axes[1][1].plot(x,y)
axes[1][1].set_title('Title 1 1')

fig.suptitle("Multiple Plots", fontsize=18)

fig.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

for ax in axes.flatten():
    ax.plot(x,y)
```
