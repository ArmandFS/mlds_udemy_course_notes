```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#create seaborn gridplots
#call a catplot:
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

```python
#add columns to catplot:
sns.catplot(data=df, x='gender', y='math score', kind='box', row='test preparation course', col='lunch')
```

```python
#create catplot but with violinplots
sns.catplot(data=df, x='gender', y='math score', kind='violin', row='test preparation course', col='lunch')
```

```python
#paigrid functionality
g = sns.PairGrid(data=df, hue='race/ethnicity')
#continue the paigrid and map it:
#fill in the uppercorner with whatever kind of plot
g = g.map_upper(sns.scatterplot)
#now fill in the lower corner
g = g.map_lower(sns.kdeplot)
#fill the diagplot OR kde
g = g.map_diag(sns.kdeplot)
#add legend
g = g.add_legend()
```

```python
#same plots but with different palette:
#paigrid functionality
g = sns.PairGrid(data=df, hue='race/ethnicity',palette='rocket')
#continue the paigrid and map it:
#fill in the uppercorner with whatever kind of plot
g = g.map_upper(sns.scatterplot)
#now fill in the lower corner
g = g.map_lower(sns.kdeplot)
#fill the diagplot OR kde
g = g.map_diag(sns.kdeplot)
#add legend
g = g.add_legend()
```
