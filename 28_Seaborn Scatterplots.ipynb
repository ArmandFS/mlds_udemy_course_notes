### 28 - Intro to Seaborn

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df
```

```python
#visualize with seaborn
plt.figure(figsize=(12,6), dpi=100)
sns.scatterplot(x= 'salary', y='sales', data=df, hue='salary')
```

```python
#visualize with seaborn for hue and levels of education
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(x= 'salary', y='sales', data=df, hue='level of education')
```

```python
#visualize scatterplot using the pallette parameter
plt.figure(figsize=(12,6), dpi=100)
sns.scatterplot(x= 'salary', y='sales', data=df, hue='level of education', palette='inferno')
```

```python
#using the alpha and size parameter
plt.figure(figsize=(12,6), dpi=100)
sns.scatterplot(x= 'salary', y='sales', data=df, hue='level of education', palette='inferno', size='salary', alpha=0.8)
```

```python
#using styles in the scatterplot
#using the alpha and size parameter
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(x= 'salary', y='sales', data=df, style='level of education', palette='inferno',  hue='level of education')
```
