```python
#test missing values on dataset
df = pd.read_csv('movie_scores.csv')
df
```

```python
#check if df is null
df.isnull()
```

```python
df[df['pre_movie_score'].isnull() & (df['first_name'].notnull())]
```

```python
#Drop data
#dropna will dorp any rows that have missing values
df.dropna()
```

```python
#drop with threshold
df.dropna(thresh=1)
```

```python
df.dropna(subset=['last_name'])
```

```python
#fill missing data
df.fillna('New Value!')
```

```python
#fill missing data with zeros
df['pre_movie_score'] = df['pre_movie_score'].fillna(0)
df
```

```python
#fill missing values with mean 
df['post_movie_score'].fillna(df['pre_movie_score'].mean())
#make this change permanent
df['post_movie_score'] = df['post_movie_score'].fillna(df['pre_movie_score'].mean())
df
```

```python
#clean up entire data
df = df.dropna()
df
```
