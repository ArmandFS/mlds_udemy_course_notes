### 21 - Matplotlib Intro and Basics

```python
#import the matplotlib library and other ds libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#create array
x = np.arange(0,20)
y = 2 * x
y
```

```python
#plot the array
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,6)
plt.ylim(0, 15)
plt.title('Simple X and Y Array Plot')
#savefig code
#plt.savefig('xplot.png')
```
