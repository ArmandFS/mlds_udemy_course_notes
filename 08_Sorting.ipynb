### Sorting and Statistical Information

```python
import numpy as np
import pandas as pd
```

```python
#read dataset and describe
df = pd.read_csv('tips.csv')
df.describe().transpose()
```

```python
#sorting the dataframe
df.sort_values('tip')
```

```python
#sort multiple columns
df.sort_values(['total_bill','tip', 'size', 'price_per_person'])
```

```python
#find max values
df['total_bill'].max()
```

```python
#find index max/min values
df['total_bill'].idxmin()
```

```python
#correlation check
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.corr()
```

```python
#categorical columns
df['sex'].value_counts()
```

```python
df['day'].unique()
```

#### Replacing Values

```python
#df head to see current dataframe
df.head()
```

```python
#replace a single value
df['sex'].replace('Female','F')
```

```python
#replace multiple values
df['sex'].replace(['Female', 'Male'], ['F', 'M'])
```

```python
#replacing values using map
mymap = {'Female': 'F', 'Male':'M'}
df['sex'].map(mymap)
```

```python
#how to deal with duplicated rows
df.duplicated()
```

```python
#simple dataframe to check for dup rows
simple_df = pd.DataFrame([1,2,2], ['a', 'b', 'c'])
simple_df
```

```python
simple_df.duplicated()
```

```python
#drop duplicates
simple_df.drop_duplicates()
```

#### Inbetween Method

```python
df['total_bill']
```

```python
df['total_bill'].between(10, 20, inclusive='both')
```

```python
#nlargest and nsmallest
#find rows with the largest tip
df.nlargest(10, 'tip')
```

```python
df.nsmallest(5, 'price_per_person')
```

```python
#sample method to get random rows or columns of data
df.sample(5)
```

```python
#sample fraction to grab 10%
df.sample(frac=0.1)
```
