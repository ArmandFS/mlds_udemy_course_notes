### 29 - Distribution Plots

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Use same dataset
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
#distribution of salaries using rugplot
sns.rugplot(x='salary', height= 0.4, data=df)
```

```python
#do not use distplot which is deprecated.
sns.set(style='darkgrid')
sns.displot(data=df, x='salary', bins=50, color='green', edgecolor='black')
```

```python
#hisplot function
sns.histplot(data=df, x='salary')
```

```python
#implement KDE and rugplot
sns.displot(data=df, x='salary', bins=40, edgecolor='black', kde='True', rug=True)
```

```python
#implement only KDE
sns.kdeplot(data=df, x='salary')
```

```python
#recreate using random data.
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
#make it into a dataframe.
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
sample_ages
```

```python
#create rugplot
sns.rugplot(data=sample_ages, x='age', height=0.4)
```

```python
#create displot
sns.displot(data=sample_ages, x='age', bins=40)
```

```python
#implement KDE
sns.displot(data=sample_ages, x='age', bins=40, kde='True')
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.02)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.6, fill=True)
```
