### 32 - Comparison Plots

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#create jointplot
#relationship between reading/writing/math score.
sns.jointplot(data=df, x='math score', y='reading score')
```

```python
#create jointplot with hexagons
#relationship between reading/writing/math score.
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
#relationship between reading/writing/math score.
sns.jointplot(data=df, x='math score', y='reading score', kind='hist')
```

```python
#show kde in jointplot:
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', fill=True)
```

```python
#adding hues in jointplots
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
#pairplots
sns.pairplot(data=df)
```

```python
#pairplots with hues
sns.pairplot(data=df,hue='gender')
```

```python
#get rid of duplicates using corner parameter:
sns.pairplot(data=df,hue='gender', corner=True)
```
