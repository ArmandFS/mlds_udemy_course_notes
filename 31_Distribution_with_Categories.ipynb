### 31 - Distribution with Categories

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
#create the boxplot, focus on distribution of math scores
plt.figure(figsize=(10,4), dpi=100)
sns.boxplot(data=df, y='math score', x= 'test preparation course')
```

```python
#create the boxplot, focus on distribution of reading scores
plt.figure(figsize=(10,4), dpi=100)
sns.boxplot(data=df, y='reading score', x= 'parental level of education')
```

```python
#create the boxplot, focus on distribution of reading scores
plt.figure(figsize=(10,4), dpi=100)
sns.boxplot(data=df, y='reading score', x= 'parental level of education', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
#create the boxplot, focus on distribution of reading scores and switch the y axis
plt.figure(figsize=(10,8), dpi=100)
sns.boxplot(data=df, x='reading score', y= 'parental level of education', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
#plot a violin plot:

plt.figure(figsize=(8,11), dpi=80)
sns.violinplot(data=df, x='reading score', y= 'parental level of education')
```

```python
#create a violinplot with hue
plt.figure(figsize=(9, 12), dpi=80)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', hue= 'test preparation course', palette='Set2')
plt.legend(bbox_to_anchor=(1.25, 0.5))
```

```python
#create a violinplot with hue with splitting equal to True
plt.figure(figsize=(9, 12), dpi=90)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', hue= 'test preparation course',  split=True)
plt.legend(bbox_to_anchor=(1.25, 0.5))
```

```python
plt.figure(figsize=(9, 12), dpi=90)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', split=True, inner= 'quartile')
```

```python
#stick value for inner parameter:
plt.figure(figsize=(9, 12), dpi=90)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', split=True, inner= 'stick')
```

```python
#Bandwidth parameter
plt.figure(figsize=(9, 12), dpi=90)
sns.violinplot(data=df, x='reading score', y= 'parental level of education', split=True, inner= 'quartile', bw=0.2)
```

```python
#create swarmplot
plt.figure(figsize=(8,4), dpi=100)
sns.swarmplot(data=df, x='math score', size=7)
```

```python
#create swarmplot with smaller dot size
plt.figure(figsize=(8,4), dpi=100)
sns.swarmplot(data=df, x='math score', size=4)
```

```python
#separate by gender and add hue
#create swarmplot with smaller dot size
plt.figure(figsize=(8,4), dpi=150)
sns.swarmplot(data=df, x='math score', y='gender', size=3, hue='test preparation course')

plt.legend(bbox_to_anchor=(1.25, 0.5))
```

```python
#add dodge parameter
plt.figure(figsize=(8,4), dpi=150)
sns.swarmplot(data=df, x='math score', y='gender', size=3, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.25, 0.5))
```

```python
#make swarmplot vertical:
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',y='math score',data=df,hue='gender')
```

```python
#boxenplot example
plt.figure(figsize=(8,4), dpi=150)
sns.boxenplot(x='math score',y='gender',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='race/ethnicity',y='math score',data=df,hue='gender')
```
