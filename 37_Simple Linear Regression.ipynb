### 37 - Simple Linear Regression in Code

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
df = pd.read_csv("Advertising.csv")
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
df.head()
```

```python
plt.figure(figsize=(10,5), dpi=120)
sns.scatterplot(data=df, x='total_spend', y='sales')
```

```python
#plot the regression line:
plt.figure(figsize=(10,5), dpi=120)
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
#define data and target values.
x = df['total_spend']
y = df['sales']
```

```python
#recall y hat function
# y = mx+b
#y = B1x + B0
```

```python
#utilize np.polyfit function
np.polyfit(x,y, deg=1)
```

```python
#array returns b1 as 0.04868788 and b0 4.24302822
```

```python
#create potential spend new values
potential_spend = np.linspace(0, 500, 100)
predicted_sales = 0.04868788*potential_spend + 4.24302822
```

```python
#now we get this regression line plotted over the scatterplot of the dataset
plt.figure(figsize=(10,5), dpi=120)
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(potential_spend, predicted_sales, color='red')
```

```python
#find new spend x values based on new potential data
spend = 200
predicted_sales = 0.04868788*spend + 4.24302822
predicted_sales
```

```python
#further considerations
np.polyfit(x, y, 3)
```

```python
#we solve for the polynomial equation:
#y = B3x**3 + B2*x**2 + B1x + B0
```

```python
pot_spend = np.linspace(0, 500, 100)
pred_sales = 3.07615033e-07*pot_spend**3 + -1.89392449e-04*pot_spend**2 + 8.20886302e-02*pot_spend + 2.70495053e+00
```

```python
plt.figure(figsize=(10,5), dpi=120)
sns.scatterplot(x='total_spend', y='sales', data=df)
plt.plot(pot_spend, pred_sales, color='red')
```
