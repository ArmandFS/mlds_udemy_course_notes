### 41 - Dealing with Outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#function to create random age data
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    
    #set a random seed in the same cell as the random call
    #set seed to 42
    np.random.seed(seed)
    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)
    
    return sample_ages
```

```python
sample = create_ages()
```

```python
sample
```

```python
#let's try to visualize the distribution
sns.displot(sample, bins=20)
```

```python
sns.boxplot(sample, orient='h')
```

```python
ser = pd.Series(sample)
```

```python
ser.describe()
```

```python
#calculate IQR 
IQR = 55.25 - 42.0
lower_limit = 42.0 - 1.5*(IQR)
lower_limit
```

```python
#use numpy's percentile  function for an IQR
q75, q25 = np.percentile(sample, [75,25])
q75
```
