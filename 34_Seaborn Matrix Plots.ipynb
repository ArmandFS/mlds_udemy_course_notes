```python
#set index to pass in heatmap into dataframe
df = pd.read_csv('country_table.csv', header=0)
df = df.set_index("Countries")
df
```

```python
plt.figure(dpi=150)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True)
```

```python
#change color mapping for heatmap
plt.figure(dpi=150)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis')
```

```python
#customize color bar with reverse cmap colouring
plt.figure(dpi=150)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis_r')
```

```python
#create a cluster matrix map
plt.figure(dpi=150)
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis')
```

```python
#clustermap without columns
plt.figure(dpi=150)
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', col_cluster=False)
```
